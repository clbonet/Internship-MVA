```python
from utils_sbm import *
from utils_clustering1d import *
from clustering1d import *

import time

import numpy as np

from sklearn.metrics import adjusted_rand_score
from sklearn.cluster import k_means

from sknetwork.visualization import svg_graph
from IPython.display import SVG
```

## Communities

```python
P = get_p(2,0.8,0.02)    

adj_sbm,lab_sbm = create_graph(P,[100,100])

start = time.time()
fs = getIterateAverages(adj_sbm,plot=False,d=1,K=2,epsilon=1e-5)
centroid,labels_k,inertia = k_means(fs[-1],2)
print("t:",time.time()-start)
print("ARI K-Means 1D:",adjusted_rand_score(lab_sbm,labels_k))

plot_iterations(np.array(fs)[[0,1,2,3,4,5]],lab_sbm,inds=[0,1,2,3,4,5])
```

## Hub

```python
P = np.ones((4,4))*0.1
for i in range(3):
    P[i,i] = 0.8
    P[3,i] = 0.8
    P[i,3] = 0.8

adj_sbm,lab_sbm = create_graph(P,[15]*4)

image = svg_graph(adj_sbm, labels=lab_sbm)
SVG(image)
```
