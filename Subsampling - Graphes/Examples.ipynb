```python
import time

from sknetwork.visualization import svg_graph
from IPython.display import SVG

from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics import adjusted_rand_score

from utils_sbm import *
from subsampling import *
from kernels import *

import vem
import ocem
```

## Graphes

```python
P = get_p(2,0.8,0.02)    

adjacency,labels = create_graph(P,[50,50])
```

```python
image = svg_graph(adjacency, labels=labels)
SVG(image)
```

## Subsampling

### GRF

```python
start = time.time()
subsampler = Subsampler(kernel_adj_dist)
subsampler.fit(adjacency)
inds = subsampler.sample()
print("t",time.time()-start)
```

```python
image = svg_graph(adjacency[np.ix_(inds,inds)], labels=labels[inds])
SVG(image)
```

```python
image = svg_graph(adjacency, scores=subsampler.f)
SVG(image)
```

### GRFAll

```python
start = time.time()
subsampler = Subsampler(kernel_adj_dist)
subsampler.fit2(adjacency)
inds = subsampler.sample()
print("t",time.time()-start)
```

```python
image = svg_graph(adjacency[np.ix_(inds,inds)], labels=labels[inds])
SVG(image)
```

```python
image = svg_graph(adjacency, scores=subsampler.f)
SVG(image)
```

### GRFR

```python
start = time.time()
subsampler = Subsampler(kernel_adj_dist,euclidean_distances)
subsampler.fit(adjacency)
inds = subsampler.sample2()
print("t",time.time()-start)
```

```python
image = svg_graph(adjacency[np.ix_(inds,inds)], labels=labels[inds])
SVG(image)
```

```python
image = svg_graph(adjacency, scores=subsampler.f)
SVG(image)
```

## EM

### VEM

```python
em = vem.VEM(3)

start = time.time()
labels_vem = em.fit(adjacency,plot=False)
print("t",time.time()-start)

print("ARI",adjusted_rand_score(labels,labels_vem))
```

```python
em = vem.VEM(3)

start = time.time()
labels_vem = em.fit(adjacency,plot=True)
print("t",time.time()-start)

print("ARI",adjusted_rand_score(labels,labels_vem))
```

### OCEM

```python
inds = np.random.choice(adjacency.shape[0],replace=False,size=adjacency.shape[0]//10)

start = time.time()
em = ocem.OCEM(3)
labels_ocem = em.fit(adjacency,inds)
print("t",time.time()-start)

print("ARI",adjusted_rand_score(labels,labels_ocem))
```

```python
start = time.time()
subsampler = Subsampler(kernel_adj_dist)
subsampler.fit(adjacency,1,1/2)
inds = subsampler.sample(0.1)
print("t subsampling",time.time()-start)

start = time.time()
em = ocem.OCEM(3)
labels_ocem = em.fit(adjacency,inds)
print("t",time.time()-start)

print("ARI",adjusted_rand_score(labels,labels_ocem))
```
